```python
import numpy as np
import pandas as pd
```

```python
from matplotlib import pyplot as plt
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from collections import defaultdict
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
data1 = pd.read_csv('brand1_nflx.csv')
data1.head()
```

```python
df1 = data1[['text']]
df1.head()
```

```python
def score_speaker(speaker_df):
    df_dict = {}
    for i,response in speaker_df.iterrows():
        scorer = SentimentIntensityAnalyzer()
        scores = scorer.polarity_scores(response.text)
        df_dict[i] = scores
    df = pd.DataFrame.from_dict(df_dict)
    df = df.T
    return df
```

```python
df_sent = score_speaker(df1)
```

```python
df_sent['user']=data1['screen_name']
```

```python
df_sent.head()
```

```python
df_sent.to_csv('sentiment.csv')
```

```python
df_group = df_sent.groupby(by='user').mean()
df_group.reset_index(inplace=True)
df_group.head()
```

```python
df_group.sort_values(by='compound', inplace=True)
```

```python
df_group.head()
```

```python
#Top users with positive tweets
df_group.tail(10)
```

```python
bins = [-1,0,0.4,1]
labels = ['detractors', 'neutrals','promoters']
df_group['NPS'] = pd.cut(df_group['compound'], bins=bins, labels=labels)
df_group['NPS'].value_counts()
```
